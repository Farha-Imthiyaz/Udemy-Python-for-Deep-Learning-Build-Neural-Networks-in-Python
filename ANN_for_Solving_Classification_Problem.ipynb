```python
# Importing All the Libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### 1. Importing All the Libraries & Reading Dataset

```python
# Reading the CSV
dataset = pd.read_csv('Churn_Modelling.csv') 
dataset
```

### 2. Splitting Dataset into dependent(y) and independent(x)

```python
# Splitting dataset into dependent and independent
X = dataset.iloc[:,3:13]
Y = dataset.iloc[:,13]
```

```python
# Printing independent variables
print(X)
```

```python
# Printing dependent variables
print(Y)
```

### 3. Changing Categorical data into numericals using encording method

```python
# Importing Encoding libraries
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

#### 3.1 Using Label Encoding

```python
# Lebel Encoding for Gender column 
labelencoder = LabelEncoder()
X['Gender'] = labelencoder.fit_transform(X['Gender'])
print(X)
```

#### 3.2 Using One Hot Encoding

```python
# One hot encoding for Geography column
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('ohe', OneHotEncoder(), [1])], remainder='passthrough')
X = np.array(ct.fit_transform(X), dtype = np.str) #.toarray()
print(X)
```

```python
X = X[:, 1:]
print(X)
```

### 4. Splitting data into train and test

```python
# Splitting data into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y,
                                    test_size = 0.2, random_state = 0)
```

### 5. Features Scaling 

```python
# Scaling features
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
import pandas as pd
pd.DataFrame(X_train)
```

### 6. Creating ANN Model

```python
# Importing all libraries

from tensorflow.keras.models import Sequential 
#from keras.models import Sequential
from keras.layers import Dense


classifier = Sequential()
```

```python
# Input layer and hidden layer
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11))
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
```

```python
# Output layer
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
```

```python
# Compiling ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
# ANN fitting
classifier.fit(X_train, y_train, batch_size = 10, epochs = 100)
```

### 7. Predicting the test results

```python
# Checking the prediction
y_pred = classifier.predict(X_test)
print(y_pred)
```

```python
# More than 0.5 equal to 1 else 0
y_pred = (y_pred > 0.5)
print(y_pred)
```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```
